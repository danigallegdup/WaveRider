```python
#!/usr/bin/env python
# coding: utf-8

import librosa 
import mir_eval 
import numpy as np
import scipy as scp
import io
```

```python
class module(object):
    def __init__(self):
        self.frame_length = 2048  
        self.hop_length = 512

    def track_beats(self,audio_data):
        audio_file = io.BytesIO(audio_data)
        file, srate = librosa.load(audio_file, sr=None)

        rms_envelope = librosa.feature.rms(y=file, frame_length=self.frame_length, hop_length=self.hop_length)[0]
        nrms_envelope = librosa.util.normalize(np.diff(rms_envelope))
        times = librosa.times_like(nrms_envelope, sr=srate, hop_length=self.hop_length)
        rms_est_onsets = []
        for (i,r) in enumerate(nrms_envelope): 
           if r > 0.1:
                rms_est_onsets.append(times[i])

        new_rms_est_onsets = []
        i = 0
        n = len(rms_est_onsets)
        while i < n:
            k = i+1
            current = rms_est_onsets[i]
            new_rms_est_onsets.append(current)
            
            while k < n and (rms_est_onsets[k] - rms_est_onsets[i] < 0.1):
                k = k + 1
        
            if k < n:
                new_rms_est_onsets.append(rms_est_onsets[k])
            i = k
        unique_values = np.unique(new_rms_est_onsets)
        return unique_values
```

```python
modul = module()
```

```python
with open('src/chi.wav', 'rb') as f:
    audio_data = f.read()
```

```python
modul.track_beats(audio_data)
```
